```python
# Importar as bibliotecas necessárias
from selenium import webdriver  # biblioteca para automatização de navegador web
from selenium.webdriver.chrome.service import Service  # biblioteca para controlar o serviço do ChromeDriver
from webdriver_manager.chrome import ChromeDriverManager  # biblioteca para gerenciar o driver do Google Chrome
import time  # biblioteca para lidar com pausas/tempo
import pandas as pd  # biblioteca para trabalhar com dados em formato tabular
```

```python
# Configurar o serviço do ChromeDriver e iniciar o navegador
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Definir a URL do site que será acessado
url = 'https://www.etf.com/etfanalytics/etf-finder'

# Acessar o site
driver.get(url)
```

```python
# Pausar a execução por 5 segundos (opcional)
time.sleep(5)

# Encontrar o botão desejado usando XPath
botao_100 = driver.find_element("xpath", '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[1]/div/div[4]/button/label/span''')

# Clicar no botão usando JavaScript
driver.execute_script('arguments[0].click();', botao_100)
```

```python
# Pausar a execução por 5 segundos (opcional)
time.sleep(5)

# Encontrar o elemento que contém o número total de páginas usando XPath
n_paginas = driver.find_element('xpath', '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/label[2]''')

# Extrair o número total de páginas a partir do texto do elemento encontrado
n_paginas = n_paginas.text.replace('of', ' ')
n_paginas = int(n_paginas)
```

```python
# Criar uma lista para armazenar todas as tabelas coletadas
lista_tabelas = []

# Iterar sobre cada página de resultados
for pagina in range(0, n_paginas):

    # Encontrar a tabela de resultados usando XPath
    tabela = driver.find_element("xpath", '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/div/table''')

    # Extrair o código HTML da tabela encontrada
    tabela_html = tabela.get_attribute('outerHTML')

    # Converter a tabela HTML em um DataFrame usando a função `read_html` do pandas
    tabela_final = pd.read_html(tabela_html)[0]

    # Adicionar a tabela convertida à lista de tabelas
    lista_tabelas.append(tabela_final)
    
    # Avançar para a próxima página de resultados
    botao_avancar_pagina = driver.find_element("xpath", '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/span[2]''')
    driver.execute_script('arguments[0].click();', botao_avancar_pagina)

# Juntar todas as tabelas coletadas em um único DataFrame usando a função `concat` do pandas
tabela_completa = pd.concat(lista_tabelas)
 
```

```python
# Encontrar o botão "performance" usando XPath
botao_aba = driver.find_element("xpath", '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/ul/li[2]/span''')

# Clicar no botão "performance"
driver.execute_script('arguments[0].click();', botao_aba)

# Iterar sobre cada página de resultados
for pagina in range(0, n_paginas):

    # Encontrar o botão "voltar" para retornar à página anterior
    botao_voltar = driver.find_element("xpath", '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/span[1]''')

    # Clicar no botão "voltar"
    driver.execute_script('arguments[0].click();', botao_voltar)

```

```python
# Cria uma lista vazia para armazenar as tabelas
lista_tabelas= []

# Loop através das páginas de resultados
for pagina in range(0, n_paginas):

    # Encontra a tabela de dados na página atual usando xpath
    tabela= driver.find_element("xpath", '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/div/table''')

    # Extrai o HTML da tabela encontrada
    tabela_html= tabela.get_attribute('outerHTML')

    # Usa a biblioteca pandas para ler a tabela HTML e converter em um dataframe
    tabela_final= pd.read_html(tabela_html)[0]

    # Adiciona o dataframe à lista de tabelas
    lista_tabelas.append(tabela_final)
    
    # Encontra o botão "próxima página" e clica nele para avançar para a próxima página de resultados
    botao_avancar_pagina= driver.find_element("xpath", '''/html/body/div[5]/section/div/div[3]/section/div/div/div/div/div[2]/section[2]/div[2]/section[2]/div[2]/div/span[2]''')
    driver.execute_script('arguments[0].click();', botao_avancar_pagina)
    
# Concatena todos os dataframes da lista em um único dataframe
tabela_completa_Performance= pd.concat(lista_tabelas)
```
